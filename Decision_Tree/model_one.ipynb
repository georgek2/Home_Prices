```python
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# Reading the data
data = pd.read_csv('train.csv', index_col=0)

# Printing first 5 rows
data.head()
```

```python
# Checking the data types
data.dtypes
```

```python
# Reviewing more about the data
data.info()
```

```python
# Checking the missing values per column
data.isnull().sum()
```

```python
# A list of all the feature names
data.columns
```

```python
# A list of all the categorical features
obj_cols = [col for col in data.columns if data[col].dtype == 'object']

obj_cols
```

```python
# A slice of the data
# Only the object columns
data[obj_cols].head()
```

```python
# Selecting the features
# Various ways to do this

# Selecting only the numerical columns
# By dropping the categorical variables
X = data.drop(obj_cols, axis=1)

X.head()
```

```python
X.dtypes
```

```python
# Excluding the objects directly
X = data.select_dtypes(exclude='object')

X.head()
```

```python
# Creating a list comprehension of numerical columns
num_cols = [col for col in data.columns if data[col].dtype in ['int64', 'float64']]

num_cols
```

```python
# Then select the numerical columns only
X = data[num_cols]

X.head()
```

```python
X.isnull().sum()
```

```python
# Dropping all the columns that have missing entries
X = X.dropna(axis=1)

X.isnull().sum()
```

```python
# Setting the target
y = X.SalePrice

# Separating it from the features
X = X.drop(['SalePrice'], axis=1)

X.head()
```

```python
# Splitting the training and the validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
y_train.shape
```

```python
# Setting the model
model = DecisionTreeRegressor(random_state=0)
```

```python
# Training the model
model.fit(X_train, y_train)
```

```python
# Making predictions using the fitted model
predictions = model.predict(X_valid)

predictions
```

```python
len(predictions)
```

```python
X_valid.head()
```

```python
y_valid.head()
```

```python
# Evaluating the model's performance
mae = mean_absolute_error(predictions, y_valid)

mae
```
